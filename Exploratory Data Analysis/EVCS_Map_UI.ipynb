```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, silhouette_score
import matplotlib.pyplot as plt

# Read the CSV data
df = pd.read_csv('CS_Processed_v3.csv')

# Convert 'Total Duration' and 'Charging Time' to timedelta
df['Total Duration'] = pd.to_timedelta(df['Total Duration'])
df['Charging Time'] = pd.to_timedelta(df['Charging Time'])

# Extract hours from 'Total Duration' and 'Charging Time'
df['Total Duration (hours)'] = df['Total Duration'].dt.total_seconds() / 3600
df['Charging Time (hours)'] = df['Charging Time'].dt.total_seconds() / 3600

# Extract year from 'End Date'
df['Year'] = pd.to_datetime(df['End Date'], format='mixed', errors='coerce').dt.year

# Group by zip code, charging station and aggregate
grouped_df = df.groupby(['ZipCode', 'Charging Station']).agg({
    'Charging Ports': 'max',
    'Total Duration (hours)': 'sum',
    'Charging Time (hours)': 'sum',
    'Energy Consumed': 'sum'
}).reset_index()

# Calculate utilization rate
grouped_df['Utilization Rate'] = grouped_df['Charging Time (hours)'] / grouped_df['Total Duration (hours)']

# Normalize data
scaler = StandardScaler()
X = scaler.fit_transform(grouped_df[['Charging Ports', 'Utilization Rate', 'Energy Consumed']])

# Cluster with K-means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
grouped_df['Cluster'] = kmeans.labels_

# Calculate silhouette score
silhouette_avg = silhouette_score(X, kmeans.labels_)
print(f"Silhouette Score: {silhouette_avg:.2f}")

# Predict ports for 2024 and 2025 with 20% increase each year
grouped_df['Ports 2024'] = grouped_df['Charging Ports'] * 1.2
grouped_df['Ports 2025'] = grouped_df['Ports 2024'] * 1.2

# Round the predicted ports to the nearest integer
grouped_df['Ports 2024'] = grouped_df['Ports 2024'].round().astype(int)
grouped_df['Ports 2025'] = grouped_df['Ports 2025'].round().astype(int)

# Create a new DataFrame with the desired columns
output_df = grouped_df[['ZipCode', 'Charging Station', 'Charging Ports', 'Ports 2024', 'Ports 2025']]

# Rename columns
output_df.columns = ['Zip Code', 'Charging Station', 'Existing Charging Points', 'Predicted Charging Points 2024', 'Predicted Charging Points 2025']

# Save the DataFrame to a CSV file
output_df.to_csv('charging_points_forecast.csv', index=False)

print("CSV file 'charging_points_forecast.csv' has been created.")
```

```python
import pandas as pd
import folium

# Read the CSV files
existingstations_unique = pd.read_csv('existing_charging_stations.csv')
existingstations_alt_fuel = pd.read_csv('Alt_Fuel_Palo_Alto_All_Existing_Stations.csv', encoding='latin1')
new_stations_df = pd.read_csv('new_charging_stations.csv')
charging_points_forecast = pd.read_csv('charging_points_forecast.csv')

# Group by 'Charging Station' and find the max latitude and longitude for each group
result = existingstations_unique.groupby('Charging Station').agg({'ZipCode': 'first','Latitude': 'max', 'Longitude': 'max'}).reset_index()

# Print the resulting DataFrame
print(result)

result.to_csv("existing_charging_stations.csv",index=False)

# Select the desired columns from existingstations_unique
columns_to_keep = ["Station Name", "Street Address", "ZIP", "Facility Type", "Latitude", "Longitude", "EV Connector Types", "EV Level2 EVSE Num"]
existingstations_alt_fuel = existingstations_alt_fuel[columns_to_keep]

# Get user input for the zip code
zip_code = input("Enter a zip code: ")

# Filter the existing and new charging stations based on the user-input zip code
existing_stations_filtered = existingstations_unique[existingstations_unique['ZipCode'] == int(zip_code)]
existing_stations_alt_fuel_filtered = existingstations_alt_fuel[existingstations_alt_fuel['ZIP'] == int(zip_code)]
new_stations_filtered = new_stations_df[new_stations_df['Zip Code'] == float(zip_code)]

# Create a map centered on the mean latitude and longitude of the filtered existing stations
if not existing_stations_filtered.empty:
    map_center = [existing_stations_filtered['Latitude'].mean(), existing_stations_filtered['Longitude'].mean()]
else:
    map_center = [new_stations_filtered['Latitude'].mean(), new_stations_filtered['Longitude'].mean()]

map_zoom = 12
m = folium.Map(location=map_center, zoom_start=map_zoom)

# Plot the filtered existing charging stations on the map
for _, row in existing_stations_filtered.iterrows():
    # Get the corresponding row from charging_points_forecast based on Zip Code and Charging Station
    forecast_row = charging_points_forecast[(charging_points_forecast['Zip Code'] == row['ZipCode']) & (charging_points_forecast['Charging Station'] == row['Charging Station'])]

    if not forecast_row.empty:
        popup_text = f"<div style='width: 300px;'><strong>Existing Station:</strong> {row['Charging Station']}<br>" \
                     f"<strong>Zip Code:</strong> {row['ZipCode']}<br>" \
                     f"<strong>Existing Charging Points:</strong> {forecast_row['Existing Charging Points'].values[0]}<br>" \
                     f"<strong>Predicted Charging Points 2024:</strong> {forecast_row['Predicted Charging Points 2024'].values[0]}<br>" \
                     f"<strong>Predicted Charging Points 2025:</strong> {forecast_row['Predicted Charging Points 2025'].values[0]}</div>"
    else:
        popup_text = f"<div style='width: 300px;'><strong>Existing Station:</strong> {row['Charging Station']}<br>" \
                     f"<strong>Zip Code:</strong> {row['ZipCode']}</div>"

    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(color='green', icon='bolt', prefix='fa')
    ).add_to(m)

# Plot the filtered existing charging stations in alt_fuel dataset on the map
for _, row in existing_stations_alt_fuel_filtered.iterrows():
    popup_text = f"<div style='width: 300px;'><strong>Station Name:</strong> {row['Station Name']}<br>" \
                 f"<strong>Street Address:</strong> {row['Street Address']}<br>" \
                 f"<strong>Zip Code:</strong> {row['ZIP']}<br>" \
                 f"<strong>Facility Type:</strong> {row['Facility Type']}<br>" \
                 f"<strong>EV Connector Types:</strong> {row['EV Connector Types']}<br>" \
                 f"<strong>EV Level2 EVSE Num:</strong> {row['EV Level2 EVSE Num']}</div>"

    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(color='green', icon='bolt', prefix='fa')
    ).add_to(m)

# Plot the filtered new charging stations on the map
for _, row in new_stations_filtered.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"<div style='width: 300px;'><strong>New Station:</strong> Zip Code {int(row['Zip Code'])}</div>",
        icon=folium.Icon(color='blue', icon='bolt', prefix='fa')
    ).add_to(m)

# Add a legend to the map
legend_html = '''
     <div style="position: fixed;
                 bottom: 50px; left: 50px; width: 150px; height: 90px;
                 border:2px solid grey; z-index:9999; font-size:14px;
                 ">&nbsp; Legend <br>
                   &nbsp; <i class="fa fa-bolt" style="color:green"></i> &nbsp; Existing Stations<br>
                   &nbsp; <i class="fa fa-bolt" style="color:blue"></i> &nbsp; New Stations
      </div>
     '''
m.get_root().html.add_child(folium.Element(legend_html))

# Save the map as an HTML file
#m.save("map.html")
m
```
